```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url = ' https://www.skysports.com/premier-league-table'

r = requests.get(url)
soup = BeautifulSoup(r.text,'html.parser')
# print(soup)

league_table = soup.find('table', class_='standing-table__table callfn')
# print(league_table)
```

```python
tablelist = []

for team in league_table.find_all('tbody'):
    rows = team.find_all('tr')
    for row in rows:
        pl_no = row.find('td', class_='standing-table__cell').text.strip()
        pl_team = row.find('td', class_='standing-table__cell standing-table__cell--name').text.strip()
        #pl_points = row.find('td', class_='standing-table__cell is-hidden--bp35').text.strip()
        pl_game = row.find_all('td', class_='standing-table__cell')[2].text
        pl_win = row.find_all('td', class_='standing-table__cell')[3].text
        pl_drawn = row.find_all('td', class_='standing-table__cell')[4].text
        pl_lose = row.find_all('td', class_='standing-table__cell')[5].text
        pl_goalsFor = row.find_all('td', class_='standing-table__cell')[6].text
        pl_goalsAgainst = row.find_all('td', class_='standing-table__cell')[7].text
        pl_goalDifference = row.find_all('td', class_='standing-table__cell')[8].text
        pl_point = row.find_all('td', class_='standing-table__cell')[9].text
        #print(pl,no,pl_team,pl_game,pl_win,pl_drawn,pl_lose,pl_goalsFor,pl_goalsAgainst,pl_goalDifference,pl_point)
        
        table = {
            '#':pl_no,
            'Team':pl_team,
            'Played':pl_game,
            'Won':pl_win,
            'Lose':pl_lose,
            'GF':pl_goalsFor,
            'GA':pl_goalsAgainst,
            'GD':pl_goalDifference,
            'Points':pl_point
        }
        
        tablelist.append(table)
        #print(table)
```

```python
premier_league = pd.DataFrame(tablelist)
premier_league
```
